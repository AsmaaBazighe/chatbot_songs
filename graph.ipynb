```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("joebeachcapital/30000-spotify-songs")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os
dataset_files = [f for f in os.listdir(path) if f.endswith('.csv')]
file_path = os.path.join(path, dataset_files[0])
df = pd.read_csv(file_path)
df
```

```python
df.columns
```

```python
df = df.dropna()
```

```python
from google.colab import userdata
import os
GEMINI_API_KEY=userdata.get('GEMINI_API_KEY')
NEO4J_URI=""
NEO4J_USERNAME=""
NEO4J_PASSWORD=""
os.environ['GEMINI_API_KEY']=GEMINI_API_KEY
os.environ['NEO4J_URI']=NEO4J_URI
os.environ['NEO4J_USERNAME']=NEO4J_USERNAME
os.environ['NEO4J_PASSWORD']=NEO4J_PASSWORD
```

```python
from neo4j import GraphDatabase

# Connexion à la base Neo4j
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))
```

```python
# Fonction pour insérer les données
def insert_data(tx, query, data):
    tx.run(query, data)

# Requête Cypher pour insérer les nœuds et les relations
songs_query = """
UNWIND $songs AS song
MERGE (t:Track {id: song.track_id})
SET t.name = song.track_name,
    t.popularity = song.track_popularity,
    t.danceability = song.danceability,
    t.energy = song.energy,
    t.tempo = song.tempo,
    t.duration_ms = song.duration_ms

MERGE (a:Artist {name: song.track_artist})
MERGE (al:Album {id: song.track_album_id, name: song.track_album_name})
MERGE (p:Playlist {id: song.playlist_id, name: song.playlist_name, genre: song.playlist_genre, subgenre: song.playlist_subgenre})

MERGE (t)-[:PERFORMED_BY]->(a)
MERGE (t)-[:PART_OF]->(al)
MERGE (t)-[:INCLUDED_IN]->(p)
"""
songs_data = df.to_dict('records')
def insert_data(tx, query, data):
    tx.run(query, songs=data)


with driver.session() as session:
    session.write_transaction(insert_data, songs_query, songs_data)

print("Données insérées avec succès dans Neo4j.")

# Fermer le driver
driver.close()
```
